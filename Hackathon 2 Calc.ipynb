# Recommendations cost calculator


```python
rec = pd.read_csv('recommendations.csv')
rec.head()
```

```python
rec2 = rec[['IMPROVEMENT_ID', 'IMPROVEMENT_ID_TEXT', 'INDICATIVE_COST']]
rec2
```

```python
rec3 = rec2.drop_duplicates()
rec3
```

```python
cost = rec3['INDICATIVE_COST'].tolist()
cost
```

```python
mylist = []
for element in cost:
    mylist.append(element.rsplit('£',1)[1])
    
mylist
```

```python
max_cost = []
for element in mylist:
        max_cost.append(element.replace(',', ''))
```

```python
max_cost
```

```python
rec3['max_cost'] = max_cost
```

```python
rec3
```

```python
rec4 = rec3.set_index('IMPROVEMENT_ID_TEXT').to_dict()['max_cost']
rec4= {k.lower(): v for k, v in rec4.items()}
rec4
```

```python
def get_improvements():
    improvements = {}
    for key in ["Improvement 1", 'Improvement 2', 'Improvement 3']:
        i = input( key + ": " )
        measure = i
        improvements[ key ] = measure
    
    price = []
    for value in improvements.values():
        for imp in rec4.keys():
            if value in imp:
                price.append(int(rec4[imp]))
                
    max_price = sum(price)

    print('Individual costs (respectively):', price)
    print('Estimated Total Maximum price: £',max_price)
get_improvements()
```
